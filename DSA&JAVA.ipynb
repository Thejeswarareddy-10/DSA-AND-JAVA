# About data structure 
# 1) Binary Search Tree  <br>
* Binary search tree is a type of data structrure in which it has atmost two children known as left and right child. In which the left subtree and its children has value lesser than the parent node and right subtree and its children has values higher than the parent node.<br>
  
  
* This data structure has efficient insertion, deletion andd searching in the time-complexity of O(log n) in which n is number of nodes in that tree.<br>
**Advantages of Binary search tree over other data structures**<br>
* Searching in sorted array takes (O(log n)) time but insertion and deletion takes (O(n)) time. Whereas in BST it takes (O(log n)) for all the operations.
* Searching in a linked list is O(n) (linear search), whereas BST has O(log n) search and also BST sorts the data automatically but linked lists need seperate sorting operations.
* So overall, the operations like insertion, deletion and searching in BST takes less time and more effectively when compared to other data structures.<br>
**What  does BST do in our project :**<br>
* Each node in the BST represents a students record. The records are stored in sorted order based on their roll numbers.
* Whenever a new student result is available it automatically sorts his result into the structure without additional sorting operations.
* To search in a BST data structure it will be easier because of its properties like left child is lesser than the root node and right child is greater then the root node.
* It will also display the results in ascending order it is also called inorder traversal property of BST.<hr>
# 2) Priority Queue <br>
* A Priority Queue is a special type of queue in which each element is associated with a priority. Elements with higher priority are dequeued before elements with lower priority, regardless of the order in which they were added.
* It has different time-complexities for each operations like (O(n log n)) for insertion and sorting, (O(n)) for searching and deletion where n is number of nodes.<br>
***Advantages of Priority queue over other data structures:***<br>
* The Priority Queue is more efficient for accessing the highest or lowest priority element compared to an unsorted array.
* Priority Queue is better for dynamically managing tasks with different priorities, while a Linked List is better for ordered sequences.
* Priority Queue allows priority-based access, whereas Stacks follow LIFO (Last-In-First-Out), making stacks unsuitable for scheduling tasks where priority matters.
* A Priority queue sorts the data based on our priorities whereas a normal queue data structure follows the principle first in first out(FIFO) in whihc it doesn't have a sorted data.
* A priority queue has is faster time-complexity in insertion and deletion operations and also in priority queue data structure can delete elements from a particular position without shifting elements unlike sorted array.<br>
**What does priority queue do in our project :** <br>
* The Priority Queue in our project helps in managing students's results based on their overall percentage. It is primarily used for ranking students in descending order of their results.
* When the student's results are added, the results are sorted in descending order based on their average percentage of all subjects and the top performing students will be displayed first.
* When a student data is deleted their record is also removed from the queue this ensures the ranking stays updated from time to time.<br><hr>
